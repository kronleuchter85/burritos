```python
import psycopg2
import sys
import pprint
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datetime import time , datetime , date

import plotly.plotly as py
import plotly.graph_objs as go
import plotly.offline as off

```

```python
conn_string = "host='nitsuga.com.ar' dbname='gtrader' user='postgres' password='postgres'"
print ("Connecting to database\n	->%s" % (conn_string))
conn = psycopg2.connect(conn_string)
cursor = conn.cursor()
#cursor.close()
#conn.close()

print ("Connected!\n")
```

```python
def analyze(prod_name, df  , features ):

    data = []

    ff = go.Scatter(
        x=df.index,
        y=df[features],
        name = prod_name,
        line = dict(color = '#17BECF'),
        opacity = 0.8)
    data.append(ff)
        
    layout = dict(
        title=prod_name,
        xaxis=dict(
            rangeselector=dict(
                buttons=list([
                    dict(count=1,
                         label='1m',
                         step='month',
                         stepmode='backward'),
                    dict(count=6,
                         label='6m',
                         step='month',
                         stepmode='backward'),
                    dict(step='all')
                ])
            ),
            rangeslider=dict(),
            type='date'
        )
    )

    fig = dict(data=data, layout=layout)
    off.plot(fig , filename= prod_name + '.html')
```

```python
def get_frame(table):
    df = pd.read_sql("select * from " + table ,conn)
    print ('Leidos:',len(df))
    return df

def clean_data(frame):
    df2 = frame.copy()
    df2['datetime'] = df2.apply(lambda x : datetime.combine( x['event_date'] , x['event_time'] ) , axis = 1)
    df2['event_time'] = df2['event_time'].apply(lambda x : time.strftime(x , '%H:%M:%S.%f'))
    df2['event_date'] = df2['event_date'].apply(lambda x : date.strftime(x , '%Y-%m-%d'))
    df2 = df2.set_index(['datetime'])
    return df2
```

```python
df1 = get_frame('nymex_future_gc_201712')
df2 = clean_data(df1)
```

```python
trades_day_1 = df2[(df2['event_type'] == 'TRADE') & (df2['event_date'] == '2017-11-01')]
analyze('nymex_future_gc_201712' ,trades_day_1, 'event_price')
```

```python
df_trades = trades_day_1.copy()
df_trades['diff'] = trades_day_1['event_price'].diff() 
df_trades['diff'] = df_trades['diff'].apply(lambda x : np.round(x , decimals=1))
df_trades = df_trades.dropna()
```

```python
analyze('GC', df_trades , 'diff' )
```

```python
plt.rcParams['figure.figsize'] = (13, 8)
df_trades['diff'].head(400).plot()
plt.show()
trades_day_1['event_price'].head(400).plot()
plt.show()
```

```python
def get_label_by_diff(diff):
    
    if diff >= -0.1 and diff <= 0.1:
        return  0
    
    elif diff > 0.1:
       # print 'label= 1 , diff=',diff
        return 1
    else:
        #diff < -0.1:
        #print 'label= -1 , diff=',diff 
        return -1
    
    
df_trades['label'] = df_trades['diff'].apply(lambda x : get_label_by_diff(x) )
```
